```python
%autoreload 2
%matplotlib inline
from bancor.logic import Bancor
from bancor.graph import Graph
from bancor.debug import Debug, DummyDebug
```

```python
# 準備金残高 connector_barance
# 供給量 bnt_supply
# 固定準備率 weight
def action(connector_barance, bnt_supply, weight):
    b = Bancor(connector_barance, bnt_supply, weight, Graph, DummyDebug)
    b.buy(300.0)
    b.buy(700.0)
    b.sell(1000.0)
    b.buy(400.0)
    b.sell(200.0)
    b.sell(300.0)
    b.plot()
```

```python
action(30000.0, 200000.0, 0.2)
action(30000.0, 200000.0, 0.1)
action(10000.0, 200000.0, 0.1)
```
